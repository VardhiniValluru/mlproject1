```python
X = dataset.iloc[:,[2,5,7]].values
```

<span style="color:black"><b><h3>Using the elbow method to find the optimal number of clusters</b></h3></span>

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'pink', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'violet', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'orange', label = 'Cluster 3')


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'lavender', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

 <span style="color:black"><b><h3>Checking the accuracy of clustering</b></h3></span>

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.labels_
silhouette_avg = silhouette_score(X,labels)
print("The silhouette score is ",silhouette_avg)
```

## Data Preprocessing

```python
dataset = pd.read_csv('DATASETAGAIN2.csv', header = None)
transactions = []
for i in range(0, 1336):
    transactions.append([str(dataset.values[i,j]) for j in range(8,18)])
```

## Importing libraries

```python
from apyori import apriori
```

## training the apriori model on the dataset

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.04, 
                min_confidence = 0.1, min_lift = 2,
                min_length = 2, max_length = 2)
```

### visualizing the results

```python
results = list(rules)
```

```python
results
```

## Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
